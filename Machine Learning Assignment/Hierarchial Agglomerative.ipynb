```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
customers = pd.read_csv('./customers.csv')
```

```python
customers.shape
```

```python
customers.head()
```

```python
data = customers.iloc[:,3:5]
print(data.head())
```

```python
import scipy.cluster.hierarchy as shc

plt.figure(figsize=(11,8))
plt.title("Customers Dendograms")
dendogram = shc.dendrogram(shc.linkage(data,method='ward'))
```

```python
plt.figure(figsize=(11,8))
plt.title("Dendograms")
dendogram = shc.dendrogram(shc.linkage(data, method='ward'))
plt.axhline(y=200,color='r',linestyle='--')
```

```python
from sklearn.cluster import AgglomerativeClustering
clusters = AgglomerativeClustering(n_clusters=5,affinity='euclidean',linkage='ward')
clusters.fit_predict(data)
```

```python
plt.figure(figsize=(11,8))
plt.scatter(data['Annual Income (k$)'],data['Spending Score (1-100)'], c=clusters.labels_, cmap='rainbow')
plt.show()
```
